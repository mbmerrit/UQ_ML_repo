## Next Part: Monte Carlo basics

Monte Carlo (MC) methods are a broad class of statistical methods that rely on random number generations to sample some quantity repeatedly. We do this to get some overall idea of the statistics of some process. For example, if we know some average rate of return and volatility of a stock (from historical data), we can do many MC simulations to get an idea of the upper and lower bounds on the performance of this stock, so how high or how low can we expect the price to go?

MC methods use several different kinds of random number generators, you may have heard of the Normal distribution, the uniform distribution, etc. These are all different approaches to random number generations. We will just stick with [Normal random variables](https://en.wikipedia.org/wiki/Normal_distribution) for now. 

Here are the basic building blocks of simulating stock growth with MC. You have a **growth rate**, sometimes called the **drift**, and the **random deviations** from this rate, called a **shock**. Back to that equation:

$$\Delta S = S \times \left(\mu \Delta t + \sigma \varepsilon \sqrt{\Delta t}\right),$$ 

this is an exponential growth equation, where the rate of change is proportional to the current value. The larger the value, that faster it grows. 

```python
# Here is a fake example of stock growth:
InitialValue = 10.0
mu = .002 # some average rate of growth, could be inferred from data
sigma = .025 # this is like a volatility
time = np.arange(1000) # time goes from zero to 1000, it could be whatever
StockValues = np.zeros((time.shape[0], 1)) # an empty array of values, we will fill this in a loop
StockValues[0] = InitialValue

# Now to do a simulation. We will use a "for loop" to do this, which iterates a set number of times
for i in range(time.shape[0]-1):
    
    drift = mu * (time[i+1] - time[i])
    epsilon = rnd.normal(0.0, 1.0) # the first parameter is mean, second is standard deviation
    shock = sigma * epsilon * np.sqrt(time[i+1] - time[i])
    StockValues[i+1] = StockValues[i] + (StockValues[i] * (drift + shock))
    
```

```python
# Now we plot the data
plt.plot(time, StockValues, '.')

# You can run this simulation as many times as you want and it will be differnt every time
```

```python
# Now I'm going to define a function that runs this simulation N times and we will plot them all

def StockSim(N, mu, sigma, InitialValue):
    time = np.arange(1000) # time goes from zero to 1000, this will stay hard-wired into the function
    StockValues = np.zeros((time.shape[0], N)) # an empty array of values, we will fill this in a loop
    StockValues[0,:] = InitialValue

    for j in range(N):
        for i in range(time.shape[0]-1):
    
            drift = mu * (time[i+1] - time[i])
            epsilon = rnd.normal(0.0, 1.0) # the first parameter is mean, second is standard deviation
            shock = sigma * epsilon * np.sqrt(time[i+1] - time[i])
            # this next line is the most interesting step, Google "Euler's Method" for more info
            StockValues[i+1,j] = StockValues[i,j] + (StockValues[i,j] * (drift + shock))
        
    return StockValues, time
```

```python
N = 20 # we will do N total simulations, we just need to run the function we defined above
StockValues, time = StockSim(N, .002, .02, 10.0) # you can adjust these parameters
```

```python
# Now we visualize them all
for i in range(N):
    plt.plot(time, StockValues[:,i])
```

## Next, we will look at applying this method to historical stock data

The main idea here is that we need to determine the important parameters ($\mu, \sigma, \varepsilon$) from historical data. This is the more data analysis-related problem. Once we have determined these parameters, we can predict future values and look at confidence intervals. The first option is to solve for $\mu$ and $\sigma$ and then attempt to estimate their average values from the formulae:
$$\Delta S = S \times \left(\mu \Delta t + \sigma \varepsilon \sqrt{\Delta t}\right) \implies \mu = \frac{\Delta S}{S \Delta t } - \frac{\sigma \varepsilon}{\sqrt{\Delta t}} $$

$$\Delta S = S \times \left(\mu \Delta t + \sigma \varepsilon \sqrt{\Delta t}\right) \implies \sigma = \frac{\Delta S}{S \varepsilon\sqrt{\Delta t} } - \frac{\mu\sqrt{\Delta t}}{\varepsilon} $$
