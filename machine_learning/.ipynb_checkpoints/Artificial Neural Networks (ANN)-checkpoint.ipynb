# Introduction to Coding Neural Networks

In this demo we will cover the fundamentals of constructing and training neural networks. Additionally, we will demonstrate how to use tools (i.e. PyTorch) to make the this task faster and easier. 

We will briefly discuss how neural networks are typically presented and how we can translate this representation into something more concrete. Recall the "computational graph" illustration of a neural network, shown below. 

<img src="neural_net.png" width="500"/>

From the above we can see that this is a two-layer network (since you count the hidden layers and the output layer, i.e., the ones with tunable weights) that inputs two numbers (features), applies a five-neuron hidden layer, and then outputs one number. Concretely, for inputs $x$ and outputs $y$, the network takes the following form:

$$
y = \sigma(xW_1+b_1)W_2+b_2
$$

where $x\in\mathbb{R}^{N\times2}$, $y\in\mathbb{R}^{N\times1}$, $W_1\in\mathbb{R}^{2\times5}$, $b_1\in\mathbb{R}^{5}$, $W_2\in\mathbb{R}^{5\times1}$, $b_2\in\mathbb{R}^{1}$, and $N$ is the number of data points. These networks are typically trained using backpropagation and gradient descent which involves taking partial derivatives of the objective function with respect to the parameters of the network. To make this process more understandable, the network can be broken down into the following subsequent steps:

$$
\begin{align*}
z_1 &= xW_1+b_1\\
y_1 &= \sigma(z_1)\\
z_2 &= y_1W_2+b_2\\
r &= z_2-\hat{y}\\
L &= \frac{1}{2}\|r\|_2^2
\end{align*}
$$

where $\hat{y}$ is the output (target) data, $r$ is the residual, and $L$ is the sum of squares error. From this one can also see how to add additional hidden layers to the network. For example, consider the $i^{\text{th}}$ hidden layer, there is first a linear transformation applied to the preceding layer $y_{i-1}$ by computing $z_i = y_{i-1}W_i + b_i$. This linear transformation is then acted upon by some nonlinear activation function $f$ (e.g. the sigmoid) which yields the output of the $i^{\text{th}}$ hidden layer: $y_i = f(z_i)$.

To train the network, we want the loss $L$ to decrease toward zero. Therefore, we update each parameter $\theta$ by taking a small step $h$ in the *negative* direction of the gradient (i.e. $\theta_\text{new} = \theta - h \frac{\partial L}{\partial \theta}$). The gradients are computed analytically using chain rule. For example, for the two-layer network above, in order to compute $\partial L/\partial W_1$, we use chain rule to get

$$
\frac{\partial L}{\partial W_1} = \frac{\partial L}{\partial r}\frac{\partial r}{\partial z_2}\frac{\partial z_2}{\partial y_1}\frac{\partial y_1}{\partial z_1}\frac{\partial z_1}{\partial W_1}.
$$

As we will see, each of these partial derivatives is easily computed. **Note:** when taking derivatives of matrix vector products, order of multiplication matters, and you will need to balance the dimensions by transposing where necessary. Now that we are all experts in neural networks, we will proceed to implementation.

## Part 1: Loading and Formatting Data

First, we need data. We begin by importing some packages.

```python
import numpy as np # basically matlab
import matplotlib.pyplot as plt # for plotting
from sklearn import datasets # for dataset
```

Next, we will load everyone's favorite: the **Iris Dataset**. The inputs `x_data` are a $150\times4$ numpy array where the rows correspond to each individual data point (150 in total), and the columns correspond to the features (i.e. sepal length (cm), sepal width (cm), petal length (cm), and petal width (cm)). The outputs `y_data` contain the three classes (i.e. setosa, versicolor, and virginica), denoted in the array by `0`, `1`, or `2`.

```python
iris = datasets.load_iris() # load the data
x_data = iris.data # extract inputs
y_data = iris.target # extract outputs

# sanity check
print('x_data shape =', x_data.shape)
print('y_data shape =', y_data.shape)
```

We will preprocess the data by making two modifications. First we will only consider the first two features (i.e. columns) of the input data. Second, since neural networks make heavy use of linear algebra, we need to force the outputs to be a 2D numpy array by adding an extra dimension.

```python
x_data = x_data[:, 0:2] # only use first 2 features
y_data = y_data[:, np.newaxis] # add extra dimension

# sanity check
print('x_data shape =', x_data.shape)
print('y_data shape =', y_data.shape)
```

Let's see how the classes are distributed.

```python
# plot
plt.figure(figsize=(10, 7))
plt.scatter(x_data[:50,0], 
            x_data[:50,1], 
            c='blue',
            marker='o') # class 0
plt.scatter(x_data[50:100,0], 
            x_data[50:100,1], 
            c='green', 
            marker='^') # class 1
plt.scatter(x_data[100:,0],
            x_data[100:,1],
            c='red',
            marker='s') # class 2
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('blue = class 0, green = class 1, red = class 2')
plt.show()
```

## Part 2: Implementation using NumPy

We will proceed to build a two layer neural network, in which the user can choose the number of neurons for the hidden layer, the total number of epochs to train for, and the stepsize with which to update the weights.

We begin by defining the sigmoid function and its derivative.

```python
# sigmoid 
def sigm(x):
    return 1.0/(1.0 + np.exp(-x))

# sigmoid prime
def sigm_p(x):
    return sigm(x)*(1.0 - sigm(x))
```

Now, we need to choose some hyperparameters.

```python
epochs = 2000 # total number of epochs
neurons = 10 # number of neurons in the hidden layer
h = 0.0001 # step size for parameter update
```

We will now randomly initialize the parameters of the network using the standard normal distribution.

```python
# define network inputs and outputs
X = x_data # inputs
Y = y_data # outputs

# get dimensions
N = len(X) # number of data points
input_dim = len(X.T) # number of input features
output_dim = len(Y.T) # number of output features

# initialize network parameters
W1 = np.random.normal(size=(input_dim, neurons))
b1 = np.zeros(shape=neurons)
W2 = np.random.normal(size=(neurons, output_dim))
b2 = np.zeros(shape=output_dim)

# sanity check
print('N =',N)
print('inputs =',input_dim)
print('outputs =',output_dim)
print('W1 shape =',W1.shape)
print('b1 shape =',b1.shape)
print('W2 shape =',W2.shape)
print('b2 shape =',b2.shape)
```

The time has come, let's train! You will find that it is very important to keep track of the shapes of each component of the forward pass as well as the partial derivatives in the backward pass.

```python
# initialize empty lists
losses = [] # mean squared errors 
accuracies = [] # correct / N

# train the network
for i in range(epochs):
    
    #
    # forward pass
    #
    
    # compute the loss
    z1 = np.dot(X, W1) + b1     # [N x neurons]
    y1 = sigm(z1)               # [N x neurons]
    z2 = np.dot(y1, W2) + b2    # [N x output_dim]
    r = z2 - Y                  # [N x output_dim]
    L = np.dot(r.T, r)/2.0      # [1 x 1]
    
    #
    # store relevant info
    #
    
    # round the predictions to obtain class labels
    predict = np.round(z2).clip(0, 3)
    
    # compute the error
    error = float(np.count_nonzero(predict-Y)) / float(N)
    
    # store the accuracy
    accuracies.append(1.0 - error)
    
    # store the loss
    losses.append(np.squeeze(L))
    
    #
    # backward pass
    #
    
    # partial derivatives
    dLdr = r                    # [N x output_dim]
    drdz2 = 1                   # [N x output_dim]
    dz2dW2 = y1                 # [N x neurons]
    dz2db2 = np.ones(shape = N) # [N, ]
    dz2dy1 = W2                 # [neurons x output_dim]
    dy1dz1 = sigm_p(z1)         # [N x neurons]
    dz1dW1 = X                  # [N x input_dim]
    dz1db1 = np.ones(shape = N) # [N, ]
    
    # chain rule
    dLdW2 = np.dot(dLdr.T, dz2dW2).T
    dLdb2 = np.dot(dLdr.T, dz2db2)
    dLdW1 = np.dot((np.dot(dLdr, dz2dy1.T)*dy1dz1).T, dz1dW1).T
    dLdb1 = np.dot((np.dot(dLdr, dz2dy1.T)*dy1dz1).T, dz1db1)
    
    # gradient update
    W2 = W2 - h*dLdW2
    b2 = b2 - h*dLdb2
    W1 = W1 - h*dLdW1
    b1 = b1 - h*dLdb1
```

Let's plot the loss and accuracy to see how we did.

```python
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(121)
plt.semilogy(losses)
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Sum of Squares Error')

ax = fig.add_subplot(122)
plt.plot(accuracies,'.')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Proportion of Correct Classifications')
plt.ylim([0,1])
plt.grid()

plt.show()
```

In an effort to see what the network has "learned", we can evaluate the network on a meshgrid of the inputs and scatter the true classes on top. 

```python
# sepal length
l_min = np.floor(np.min(X[:,0]))
l_max = np.ceil(np.max(X[:,0]))

# sepal width
w_min = np.floor(np.min(X[:,1]))
w_max = np.ceil(np.max(X[:,1]))

# make a linspace
num_pts = 200
sepal_length = np.linspace(l_min,l_max,num_pts)
sepal_width = np.linspace(w_min,w_max,num_pts)

# convert to mesh
[l_mesh,w_mesh] = np.meshgrid(sepal_length,sepal_width)

# create ordered pairs
l_mesh = l_mesh.reshape(-1)[:,np.newaxis]
w_mesh = w_mesh.reshape(-1)[:,np.newaxis]
X_mesh = np.concatenate([l_mesh,w_mesh],axis=1)
y_mesh = np.round(np.dot(sigm(np.dot(X_mesh,W1)+b1),W2)+b2)

# reshape back into 2d
L_mesh = l_mesh.reshape(num_pts,num_pts)
W_mesh = w_mesh.reshape(num_pts,num_pts)
Y_mesh = np.minimum(2,np.maximum(0,y_mesh.reshape(num_pts,num_pts)))

# plot
plt.figure(figsize=(10, 7))
plt.imshow(np.flipud(Y_mesh), extent=[l_min,l_max,w_min,w_max], cmap='jet')
plt.scatter(x_data[:50,0], x_data[:50,1], c='blue', marker='o') # class 0
plt.scatter(x_data[50:100,0], x_data[50:100,1], c='green', marker='^') # class 1
plt.scatter(x_data[100:,0], x_data[100:,1], c='red', marker='s') # class 2
plt.title('blue = class 0, green = class 1, red = class 2')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.xlim(4, 8)
plt.ylim(2, 5)
plt.show()
```

There we have it, a neural network classifier with one sigmoid-activated hidden layer and linear output, using only NumPy.

## Part 3: Implementation in PyTorch

It is time to make life easier using tools. Let's first import the PyTorch module and assert that we're using CPU.

```python
import torch
device = torch.device('cpu')
```

Since we already loaded and formatted the data for the NumPy implementation above, let's just remind ourselves of what we're dealing with.

```python
print('x_data shape =', x_data.shape)
print('y_data shape =', y_data.shape)
```

Let's set the network hyperparameters again.

```python
epochs = 2000 # total number of epochs
neurons = 10 # number of neurons in the hidden layer
h = 0.1 # step size for parameter update
```

We also need to assign the dimensions.

```python
# get dimensions
N = len(x_data) # number of data points
input_dim = len(x_data.T) # number of input features
output_dim = len(y_data.T) # number of output features
```

PyTorch is different from NumPy since PyTorch objects need to carry more info (like gradients) compared to NumPy arrays. Therefore, we need to define the inputs and outputs using PyTorch tensors.

```python
# initialize inputs and outputs
X = torch.tensor(x_data, dtype=torch.float, device=device)
Y = torch.tensor(y_data, dtype=torch.float, device=device)
```

Then the network, loss function, and optimization scheme.

```python
# initialize network
MLP = torch.nn.Sequential(
    torch.nn.Linear(in_features=input_dim, out_features=neurons),
    torch.nn.Sigmoid(),
    torch.nn.Linear(in_features=neurons, out_features=output_dim))

# initialize loss function
def loss(y_pred, y_true):
    return 0.5*torch.mean((y_pred - y_true)**2)

# compile optimization scheme
parameters = MLP.parameters()
optimizer = torch.optim.SGD(parameters, lr=h)
```

We will now train the network. 

```python
# initialize empty lists
losses = [] # mean squared errors 
accuracies = [] # num. correct / N

# train for total number of epochs
for i in range(epochs):
    
    #
    # forward pass
    #
    
    # predict
    predict = MLP(X)
    
    # compute loss
    loss_value = loss(predict, Y)
    
    #
    # book keeping
    #
    
    # compute the error
    predict_np = np.round(predict.detach().cpu().numpy()).clip(0, 2)
    error = float(np.count_nonzero(predict_np - y_data)) / float(N)

    # store the accuracy
    accuracies.append(1.0 - error)

    # store the loss
    losses.append(loss_value.detach().cpu().numpy())
    
    #
    # backward pass
    #
    
    # zero out gradients from previous iteration
    optimizer.zero_grad()
    
    # apply chain rule
    loss_value.backward()
    
    # update MLP parameters
    optimizer.step()

# evaluate the trained network on a meshgrid
x_mesh = torch.tensor(X_mesh, dtype=torch.float, device=device)
y_mesh = np.round(MLP(x_mesh).detach().cpu().numpy()).clip(0, 3)
```

Well that was easy, no need to compute analytic gradients or any of the other hassles in NumPy. PyTorch gives you a good balance of control as well as ease of use. There are thousands of commands and functions you can use in PyTorch which gives you the opportunity to implement even the most complicated modern deep learning algorithms. 

Let's plot the results and see how they compare to what we did earlier using NumPy.

```python
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(121)
plt.semilogy(losses)
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Sum of Squares Error')

ax = fig.add_subplot(122)
plt.plot(accuracies,'.')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Proportion of Correct Classifications')
plt.ylim(0,1)
plt.grid()

plt.show()
```

Let's evaluate the model on a meshgrid again.

```python
# reshape y_mesh into 2d
Y_mesh = y_mesh.reshape(num_pts,num_pts).clip(0,2)

# plot
plt.figure(figsize=(10, 7))
plt.imshow(np.flipud(Y_mesh), extent=[l_min,l_max,w_min,w_max], cmap='jet')
plt.scatter(x_data[:50,0], x_data[:50,1], c='blue', marker='o') # class 0
plt.scatter(x_data[50:100,0], x_data[50:100,1], c='green', marker='^') # class 1
plt.scatter(x_data[100:,0], x_data[100:,1], c='red', marker='s') # class 2
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('blue = class 0, green = class 1, red = class 2')
plt.xlim(4, 8)
plt.ylim(2, 5)
plt.show()
```

Not bad, but we can push the accuracy higher, let's add some layers and use a better optimization scheme to see if we get improved results. To start, let's build a general MLP class that let's us customize the network architecture.

```python
class BuildMLP(torch.nn.Module):
    
    def __init__(self, layers, activation, output_activation=None):
        
        super().__init__()
        operations = []
        input_features = layers[0]
        
        # hidden layers
        for i, layer in enumerate(layers[1:-1]):
            
            operations.append(torch.nn.Linear(
                in_features=input_features,
                out_features=layer))
            input_features = layer
            operations.append(activation)
            
        # output layer
        operations.append(torch.nn.Linear(
                in_features=input_features,
                out_features=layers[-1],
                bias=True))
        if output_activation is not None:
            operations.append(output_activation)
                
        # convert module list to sequential model
        self.MLP = torch.nn.Sequential(*operations)
        
    def forward(self, x):
        
        # run the model
        y = self.MLP(x)
        
        return y
```

Now let's build a larger MLP model and train using the Adam optimizer.

```python
# import a status bar
from tqdm import tqdm

# hyperparameters
epochs = 50000
h = 0.001 

# model and optimizer
MLP = BuildMLP([2, 32, 32, 32, 1], torch.nn.LeakyReLU())

# initialize loss function
def loss(y_pred, y_true):
    return 0.5*torch.mean((y_pred - y_true)**2)

# compile optimization scheme
parameters = MLP.parameters()
optimizer = torch.optim.Adam(parameters, lr=h)

# initialize empty lists
losses = [] # mean squared errors 
accuracies = [] # num. correct / N

# train for total number of epochs
for i in tqdm(range(epochs)):
    
    # predict
    predict = MLP(X)
    
    # compute loss
    loss_value = loss(predict, Y)
    
    # compute the error
    predict_np = np.round(predict.detach().cpu().numpy()).clip(0, 2)
    error = float(np.count_nonzero(predict_np - y_data)) / float(N)

    # store the accuracy
    accuracies.append(1.0 - error)

    # store the loss
    losses.append(loss_value.detach().cpu().numpy())
    
    # zero out gradients from previous iteration
    optimizer.zero_grad()
    
    # apply chain rule
    loss_value.backward()
    
    # update MLP parameters
    optimizer.step()

# evaluate the trained network on a meshgrid
x_mesh = torch.tensor(X_mesh, dtype=torch.float, device=device)
y_mesh = np.round(MLP(x_mesh).detach().cpu().numpy()).clip(0, 3)
        
# plot the results
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(121)
plt.semilogy(losses)
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Sum of Squares Error')

ax = fig.add_subplot(122)
plt.plot(accuracies,'.')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Proportion of Correct Classifications')
plt.ylim(0,1)
plt.grid()

plt.show()
```

Let's repeat the meshgrid plot to see how the decision surface has changed.

```python
# reshape y_mesh into 2d
Y_mesh = np.minimum(2,np.maximum(0,y_mesh.reshape(num_pts,num_pts)))

# plot
plt.figure(figsize=(10, 7))
plt.imshow(np.flipud(Y_mesh), extent=[l_min,l_max,w_min,w_max], cmap='jet')
plt.colorbar()
plt.scatter(x_data[:50,0],x_data[:50,1],c='blue', marker='o') # class 0
plt.scatter(x_data[50:100,0],x_data[50:100,1],c='green', marker='^') # class 1
plt.scatter(x_data[100:,0],x_data[100:,1],c='red', marker='s') # class 2
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('blue = class 0, green = class 1, red = class 2')
plt.xlim(4, 8)
plt.ylim(2, 5)
plt.show()
```

If you ran the above stuff, you should find that while the classification accuracy had increased, from the heatmap it is clear the the network is severely overfitting. Even so, this hopefully gives you an introduction to how you can use PyTorch to solve your problems.
